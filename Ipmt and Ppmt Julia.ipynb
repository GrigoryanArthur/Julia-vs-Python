```julia
using Distributions
using StatsBase
using Dates
using NPFinancial
```

```julia
nrows=2000000
horizon_length=36
```

```julia
amount=rand(1,nrows)*10000
interest=rand(1,nrows)/10
maturities1=rand(Date(2022, 12, 31):Day(1):Date(2040, 12, 31),nrows,1)
maturities=rand(5:240, nrows)
```

```julia
loan_ipmt = Array{Float64}(undef, nrows, horizon_length)
loan_ppmt = Array{Float64}(undef, nrows, horizon_length)
```

```julia
@time begin
    for i=1:nrows
        loan_ipmt[i,:]=-ipmt.(interest[i]/12, collect(1:horizon_length), maturities[i], amount[i])
        loan_ppmt[i,:]=-ppmt.(interest[i]/12, collect(1:horizon_length), maturities[i], amount[i])
    end
end
```
